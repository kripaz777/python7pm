```python
# and & or | 
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * / ")
    if o == '+':
        print(add(a,b))
    elif o == '-':
        print(sub(a,b))
    elif o == '*':
        print(mul(a,b))
    elif o == '/' and b!=0:
        print(div(a,b))
    elif b==0:
        print("Value of b can not be zero")
    else:
        print("Invalid")
        
cal()
```
