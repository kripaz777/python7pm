```python
import csv
data = []
with open('books.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(row)

print(data)
```

```python
import csv
csv_columns = ['Sn','Names','Address','Age']
dict = data
csvfile = open("data.csv", 'w')


writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
writer.writeheader()
for dat in dict:
    writer.writerow(dat)
#     print(dat)

csvfile.close()
```
